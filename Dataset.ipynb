```python
train_path = pd.read_csv("/content/dataset_train.csv", encoding='latin-1') # ganti dengan directoy dari gcp bucket
alay_path = pd.read_csv("/content/new_kamusalay.csv", encoding='latin-1', header=None) # ganti dengan directoy dari gcp bucket
```

```python
alay_dict = alay_path.rename(columns={0: 'original', 
                                      1: 'replacement'})
```

```python
import re
def lowercase(text):
    return text.lower()

def remove_unnecessary_char(text):
    text = re.sub('\n',' ',text)    # Remove every '\n'
    text = re.sub('rt',' ',text)    # Remove every retweet symbol
    text = re.sub('user',' ',text)  # Remove every username
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ',text) # Remove every URL
    text = re.sub('  +', ' ', text) # Remove extra spaces
    return text
```

```python
def remove_nonaplhanumeric(text):
    text = re.sub('[^0-9a-zA-Z]+', ' ', text) 
    return text
```

```python
alay_dict_map = dict(zip(alay_dict['original'], alay_dict['replacement']))
def normalize_alay(text):
    return ' '.join([alay_dict_map[word] if word in alay_dict_map else word for word in text.split(' ')])
```

```python
def preprocess(text):
    text = lowercase(text) # 1
    text = remove_nonaplhanumeric(text) # 2
    text = remove_unnecessary_char(text) # 2
    text = normalize_alay(text) # 3
    return text
```

```python
train_path['Tweet'] = train_path['Tweet'].apply(preprocess)

decode_map = {0: "Positive", 1: "Negative"}
def decode_sentiment(label):
    return decode_map[int(label)]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
SEQUENCE_LENGTH=300
tokenizer = Tokenizer(num_words=2000)
tokenizer.fit_on_texts(train_path.Tweet)
import json
```

```python
def comment_to_instances(text):
    # Tokenize text
    x_test = pad_sequences(tokenizer.texts_to_sequences([text]), maxlen=SEQUENCE_LENGTH)
    json_string = json.dumps(x_test[0].tolist())
    return json.loads(json_string)
```
